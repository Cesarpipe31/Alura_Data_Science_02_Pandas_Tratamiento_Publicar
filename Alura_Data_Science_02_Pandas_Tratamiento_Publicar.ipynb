## Parte 3 **Extras**  Creando Estructuras de Datos

```python
import pandas as pd
```

### **Series**

```python
# lista data
data = [1, 2, 3, 4, 5]
```

```python
# se hace una serie a través de una lista
s = pd.Series(data)
```

```python
# Indices hechos por default, desde 0 hasta el 4
s
```

```python
# Mejorando el indice con una list comprehension
index = ['Linea' + str(i) for i in range(5)]
# Los nuevos indices serán los siguientes
index
```

```python
# A la serie ya generada, le adiciono el nuevo index
s = pd.Series(data= data, index = index)
# Nueva lista series con nuevos indices y la data inicial
s
```

### Nueva forma con un Diccionario

```python
#  Las {} indican que será un diccionario
# con la siguiente linea crearemos el indice y el valor (key and values)
data = {'Linea' + str(i): i+1 for i in range(5)}
data
```

```python
s = pd.Series(data)
s
```

```python
s1 = s + 2
s1
```

```python
s
```

```python
# Se puede hacer operaciones entre series sí y solo sí tiene exactamente el mismo indice
s2 = s + s1
s2
```

```python
data1 = {'NewIndex' + str(i): i+1 for i in range(5)}
data1
```

```python
r = pd.Series(data1)
r
```

```python
type(r)
```

```python
# esta operacion entre series NO funcionará ya que tienen diferente indicel y dará como resultado NaN
s3 = s + r
s3
```

## **DataFrame**

```python
# Una lista de listas, una matriz, con tres elementos cada uno
data = [[1,2,3], [4,5,6], [7,8,9]]
data
```

```python
# Creamos el data frame... es una matriz, filas y columnas.  El indice es el indice genérico que genera automáticamente
df1 = pd.DataFrame(data = data)
df1
```

```python
# Mejoramos el indice
index = ['Linea' + str(i) for i in range(3)]
index
```

```python
# integramos el nuevo index
df1 = pd.DataFrame(data = data, index = index)
df1
```

```python
# mejoremos los nombres de las columnas
columns = ['Columna' + str(i) for i in range(3)]
columns
```

```python
df1 = pd.DataFrame(data = data, index = index, columns = columns)
df1
```

```python
# Otra forma, generando un data frame a través de un diccionario
data = {'Columna0': {'Linea0': 1, 'Linea1': 4, 'Linea2': 7},
        'Columna1': {'Linea0': 2, 'Linea1': 5, 'Linea2': 8},
        'Columna2': {'Linea0': 3, 'Linea1': 6, 'Linea2': 9}}
data
```

```python
# generamos el data frame con el diccionario
df2 = pd.DataFrame(data)
df2
```

```python
# generemos una lista de tuplas
data2 = [(1,2,3), (4,5,6),(7,8,9)]
data
```

```python
df3 = pd.DataFrame(data, index = index, columns = columns)
df3
```

```python
# operaciones con Data Frame
df1[df1 > 0] = 'A'
df1
```

```python
# operaciones con Data Frame
df2[df2 > 0] = 'B'
df2
```

```python
# operaciones con Data Frame
df3[df3 > 0] = 'C'
df3
```

```python
#concatnemos los tres dataframes
df4 = pd.concat([df1, df2, df3])
df4
```

```python
# otra concatenacion, horizontalmente.... crecer a la derecha
# axis y concatena por el valor 1, por las filas... Linea0, Linea1, Linea2
df5 = pd.concat([df1, df2, df3], axis = 1)
df5
```

## *Parte 4. Filtrando datos*
Reporte de Análisis III
*Inmuebles Residenciales*

```python
import pandas as pd
```

```python
datos = pd.read_csv('alquiler.csv', sep = ';')
datos.head(10)
```

```python
# Creamos lista con los diferentes tipos de locales, no mostrar los repetidos, solo los únicos
list(datos['Tipo'].drop_duplicates())
```

```python
# Crearemos lista con valores que correspondan a residenciales solamente
residencial = ['Habitación',
 'Casa',
 'Departamento',
 'Casa en condominio',
 'Casa comercial',
 'Casa de villa',
 ]
```

```python
residencial
```

```python
seleccion = datos['Tipo'].isin(residencial)
seleccion
```

```python
datos.describe()
```

```python
datos_residencial = datos[seleccion]
datos_residencial
```

```python
datos_residencial.describe()
```

```python
# confirmemos que tipos existen en el nuevo data frame
list(datos_residencial['Tipo'].drop_duplicates())

```

```python
datos_residencial
```

```python
datos_residencial.shape[0]
```

```python
datos.shape[0]
```

```python
# Actulicemos el indice
datos_residencial.index = range(datos_residencial.shape[0])
datos_residencial
```
